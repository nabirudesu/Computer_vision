```python
import os
import tqdm
import cv2
import numpy as np
import matplotlib.pyplot as plt             
```

```python
img_path='/run/media/kyoraku/C08C64758C64683C/New folder/New folder/IMG_3085-01.jpeg'
def convert_im_to_px(img_path):
    image = cv2.imread(img_path)
    # Get the 3 component of RGB colors for every pixel
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 
    image=np.array(image)
    return image
```

```python
img= convert_im_to_px(img_path)
```

```python
print("Width: "+str((np.shape(img)[0]))+" pixels")
print("Height: "+str((np.shape(img)[1]))+" pixels")
print("first line of the images's pixels array : #{}" .format(img[0]))
```

```python
def display_image(img):
    plt.figure()
    plt.imshow(img)
    plt.title("Nabil Tiaiba")
    plt.show()
```
